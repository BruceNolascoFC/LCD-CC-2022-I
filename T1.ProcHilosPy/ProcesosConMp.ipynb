```python
import multiprocessing as mp
import time as t
```

## Procesos con multiprocessing

```python
# funci√≥n para calcular potencia arbitraria de numeros
def calc_potencia(num,k = 1):
  for i in num:
    print(f"Potencia {k} de {i}: {i**k}")
nums = range(1,100)
```

```python
# mostrar cuadrado
p = mp.Process(target=calc_potencia,args = (nums,2))
p.start()

p.join()
```

Mostrar cuadrados y cubos

```python
p1 = mp.Process(target=calc_potencia,args = (nums,2))
p2 = mp.Process(target=calc_potencia,args = (nums,3))
p3 = mp.Process(target=calc_potencia,args = (nums,0.5))
p1.start()
p2.start()
p3.start()
p1.join()
p2.join()
p3.join()
```

```python
# secci√≥n 1
start = t.time()
calc_potencia(nums,2)
calc_potencia(nums,3)
end = t.time()
print(f"Tiempo de ejecuci√≥n: {end-start}")
```

```python
#seccion 2
p1 = mp.Process(target=calc_potencia,args = (nums,2))
p2 = mp.Process(target=calc_potencia,args = (nums,3))

start = t.time()
p1.start()
p2.start()

p1.join()
p2.join()

end = t.time()
print(f"Tiempo de ejecuci√≥n: {end-start}")
```

```python
#seccion 3
p1 = mp.Process(target=calc_potencia,args = (nums,2))
p2 = mp.Process(target=calc_potencia,args = (nums,3))

start = t.time()
p1.start()
p1.join()
p2.start()
p2.join()

end = t.time()
print(f"Tiempo de ejecuci√≥n: {end-start}")
```
